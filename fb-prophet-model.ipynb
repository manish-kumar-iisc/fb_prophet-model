## All imports

```python
from fbprophet import Prophet
import numpy as np
import pandas as pd
```

## Reading Data

```python
train=pd.read_csv("../input/rajat-company-dataset/telemetry_sample_data_12_2020.csv")
train1=train.dropna()
```

## Preprocessing Data

```python
train1.head(30)
time=train1['timestamp']
newstamp=[]
for i in time:
    data=i[0:10]
    tme=i[11:19]
    newstamp.append(data+" "+tme)
np.shape(newstamp)
```

```python
train1['timestamp']=newstamp
train1.head(5)
train1=train1.rename(columns={"timestamp":'ds', "ts_values":'y'})
train1['ds']=train1['ds'].astype('datetime64[ns]')
# train1['time1']=pd.to_datetime(train1['time1'],format='%y%m%d')
train1.dtypes
```

```python
train1
```

## Model
https://facebook.github.io/prophet/docs/quick_start.html

```python
model=Prophet()
model.fit(train1)
```

```python
predict=["2020-11-01 00:55:00", "2020-11-01 01:55:00"]
predict=pd.DataFrame(predict).rename(columns={0:'ds'})
predict
forecast=model.predict(predict)
forecast[['yhat','trend','yhat_lower','yhat_upper']].tail()
```
